```python
b) Identifiers
   -----------
Identifiers are fundamental building blocks of a program and
are used as the general terminology for the names given to
different parts of the program viz. variables, objects,
classes, functions, lists, dictionaries etc.

Rules
-----
1) An identifier is an arbitararily long sequence of underscore,
   letters and digits.
2) The first character must be a letter or an underscore(_). 
3) The digits 0 through 9 can be part of the identifier except
   for the first character.
4) An identifier must not be a keyword of python.
5) An identifier cannot contain any special character except
   for underscore(_).
6) Identifiers are unlimited in length.  
7) Case is significant.
  Ex- valid-   Mqwq, _sjdsh8jbs, ____,etc             
      Invalid-  9hb_s, fdh_nd.sb,etc 
```
